# Using Sklearn Library

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print("Our Dataset: \n{}".format(dataset))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state=0)
```

```python
plt.scatter(X_train, y_train, marker='x', c='r')
plt.title("Total Salary")
plt.ylabel("Salary")
plt.xlabel("YearsExperience")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print("Our Test Set: {}".format(y_test))
```

```python
y_pred = regressor.predict(X_test)
print("Our Predictions: {}".format(y_pred))
```

```python
# Visualizing the Training set results
plt.scatter(X_train, y_train, marker='x', c='r', label='Actual data')
plt.plot(X_train, regressor.predict(X_train), c='b', label='Predictions')
plt.title('Salary vs Experience (Train set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

print('Linear Regression Model MSE is: ', mean_squared_error(y_test, y_pred))
print('Linear Regression Model Training score: ', regressor.score(X_train, y_train))
print('Linear Regression Model Test score: ', regressor.score(X_test, y_test))
```
